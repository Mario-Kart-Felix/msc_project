```python
import pandas as pd
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
survey = pd.read_csv('responses.csv')
```

```python
survey.describe().T
```

```python
f_oneway(survey.readability_textrank,
        survey.readability_lexrank,
        survey.readability_opinosis,
        survey.readability_human_1,
        survey.readability_human_2,
        survey.readability_human_3,
        survey.readability_human_4)
```

```python
f_oneway(survey.utility_textrank,
        survey.utility_lexrank,
        survey.utility_opinosis,
        survey.utility_human_1,
        survey.utility_human_2,
        survey.utility_human_3,
        survey.utility_human_4,
        survey.utility_wordcloud,
        survey.utility_sentiment)
```

```python
survey.boxplot(column = ['readability_textrank',
     'readability_lexrank',
       'readability_opinosis',
        'readability_human_1',
        'readability_human_2',
        'readability_human_3',
       'readability_human_4'], rot='45')
plt.xticks(ha='right')
plt.savefig('readability.png')
```

```python
from statsmodels.stats.multicomp import (pairwise_tukeyhsd,
                                         MultiComparison)

# Set up the data for comparison (creates a specialised object)
MultiComp_readability = MultiComparison(df_readability['readability'], df_readability['algorithm'])

# Show all pair-wise comparisons:

# Print the comparisons

print(MultiComp_readability.tukeyhsd().summary())
```

```python
# Set up the data for comparison (creates a specialised object)
MultiComp_utility = MultiComparison(df_utility['utility'], df_utility['algorithm'])

# Show all pair-wise comparisons:

# Print the comparisons

print(MultiComp_utility.tukeyhsd().summary())
```

```python
rank = survey.iloc[:, 12:19]
rank.columns = ['human_2', 'lexrank', 'textrank', 'opinosis', 'human_4', 'human_1', 'human_3']
```

```python
rank.boxplot(column=[ 'textrank', 'lexrank','opinosis', 'human_1', 'human_2', 'human_3', 'human_4'], rot='45')
plt.xticks(ha='right')
plt.savefig('rank.png')
```

```python
rank_compare = rank.copy()
```

```python
rank_compare['t']= 'textrank'

rank_compare['l'] = 'lexrank'

rank_compare['o'] = 'opinosis'

rank_compare['h1'] = 'human_1'

rank_compare['h2'] = 'human_2'

rank_compare['h3'] = 'human_3'

rank_compare['h4'] = 'human_4'
```

```python
rank_readability = pd.concat([rank_compare[['textrank', 't']].rename(columns={'textrank': 'rank', 't': 'algorithm'}),
                            rank_compare[['lexrank', 'l']].rename(columns={'lexrank': 'rank', 'l': 'algorithm'}),
                           rank_compare[['opinosis', 'o']].rename(columns={'opinosis': 'rank', 'o': 'algorithm'}),
                           rank_compare[['human_1', 'h1']].rename(columns={'human_1': 'rank', 'h1': 'algorithm'}),
                           rank_compare[['human_2', 'h2']].rename(columns={'human_2': 'rank', 'h2': 'algorithm'}),
                           rank_compare[['human_3', 'h3']].rename(columns={'human_3': 'rank', 'h3': 'algorithm'}),
                           rank_compare[['human_4', 'h4']].rename(columns={'human_4': 'rank', 'h4': 'algorithm'})])
```

```python
# Set up the data for comparison (creates a specialised object)
MultiComp_rank = MultiComparison(rank_readability['rank'], rank_readability['algorithm'])

# Show all pair-wise comparisons:

# Print the comparisons

print(MultiComp_rank.tukeyhsd().summary())
```
